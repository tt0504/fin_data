```python
X1 ='Total current tax assets'
X2 ='Total current tax liabilities'
X3 ='Total assets'
X4 ='Total retained earnings'
X5 ='Earnings Before Interest and Tax'
X6 ='Total equity'
X7 ='Total liabilities'
X8 ='Total income (loss) from discontinued operations'
KTB ='2809 KTB'
TCCB ='2812 T.C.C.B.'
```

```python
A_KTB = (df1.loc[KTB,X1] - df1.loc[KTB,X2])/df1.loc[KTB, X3]
```

```python
B_KTB = df1.loc[KTB,X4]/df1.loc[KTB, X3]
```

```python
C_KTB = df2.loc[KTB,X5]/df1.loc[KTB, X3]
```

```python
D_KTB = df1.loc[KTB,X6]/df1.loc[KTB, X7]
```

```python
E_KTB = df2.loc[KTB,X8]/df1.loc[KTB, X3]
```

```python
Z_score_KTB = 1.2*A_KTB + 1.4*B_KTB + 3.3*C_KTB + 0.6*D_KTB +1.0*E_KTB
```

```python
Z_score_KTB, A_KTB, B_KTB, C_KTB, D_KTB, E_KTB
```
